```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_for_V(V):
    mu_list = [0, -0.5, -1, -1.5, -2]

    fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(25, 5))

    for index, mu in enumerate(mu_list):
        with open(f'gap_plots_data/gap_size_vs_U_V_{V}_mu_{mu}.csv', 'r') as f:
            data = pd.read_csv(f, index_col=0)
        ax = axs[index]

        ax.plot(data['U'], data['delta_0'], 'x--', label='$Gr_1$', alpha=0.8)
        ax.plot(data['U'], data['delta_1'], 'x--', label='$Gr_2$', alpha=0.8)
        ax.plot(data['U'], data['delta_2'], 'x--', label='$X$', alpha=0.8)

        ax.set_ylim(bottom=0)
        ax.set_xlabel('U')
        ax.legend()
        ax.set_title(f'$V = {V}, \\mu = {mu}$')
        
    return fig
```

```python
def plot_for_mu(mu):
    V_list = [0.1, 1, 2]

    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(25, 5))

    for index, V in enumerate(V_list):
        with open(f'gap_plots_data/gap_size_vs_U_V_{V}_mu_{mu}.csv', 'r') as f:
            data = pd.read_csv(f, index_col=0)
        ax = axs[index]

        ax.plot(data['U'], data['delta_0'], 'x--', label='$Gr_1$', alpha=0.8)
        ax.plot(data['U'], data['delta_1'], 'x--', label='$Gr_2$', alpha=0.8)
        ax.plot(data['U'], data['delta_2'], 'x--', label='$X$', alpha=0.8)

        ax.set_ylim(bottom=0)
        ax.set_xlabel('U')
        ax.legend()
        ax.set_title(f'$V = {V}, \\mu = {mu}$')

    return fig
```

```python
V = 2

fig = plot_for_V(V)
fig.savefig(f'gap_plots/gap_size_vs_U_V_{V}.pdf')
```

```python
mu = 0

fig = plot_for_mu(mu)
fig.savefig(f'gap_plots/gap_size_vs_U_mu_{mu}.pdf')
```
